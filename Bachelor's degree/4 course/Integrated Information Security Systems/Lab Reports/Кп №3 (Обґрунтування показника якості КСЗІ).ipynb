#### 1. Згенеруємо випадковим чином частоту появи кожної загрози та розрахуємо імовірність появи *і*-х загроз ↓

```python
import numpy as np

# Генеруємо випадкові частоти появи кожної загрози від 1 до 10
nummer_of_threats = 10
frequency = np.random.randint(1, 11, nummer_of_threats)

print("Частота появи загроз:")
frequency
```

```python
# Обчислюємо імовірності появи кожної загрози
Pi = []
for i in range(nummer_of_threats):
    Pi.append(frequency[i] / sum(frequency))
    
print("Імовірності появи загроз:")
np.round(Pi, 2)
```

#### 2. Задамо відносні величини збитків ↓

```python
import pandas as pd

# Створимо порожній датафрейм
threat_df = pd.DataFrame()

# Заповнимо стовбці датафрейму з відповідними даними
threat_df["Назва загрози"] = ["Ненавмисне псування носіїв інформації", "Ненавмисне знищення важливої інформації", "Навмисне знищення інформації вірусами та ін.", "Затримки чи недоступність БД внаслідок некоректного квотування", "Відмова в обслуговуванні внаслідок дій зловмисника", "Підбір та вивідування паролів користувача", "Викрадення службової інформації з БД інсайдерами", "Викрадення службової інформації з файлової системи зовнішніми зловмисниками", "Відмова засобів адміністрування та контролю роботи КСЗІ", "Приховування дій зловмисника із використанням руткіта"]
threat_df["Тип загрози"] = ["Цілісність"]*3 + ["Доступність"]*2 + ["Конфеденційність"]*3 + ["Спостережність"]*2
threat_df["Імовірність загрози"] = np.round(Pi, 2)
```

```python
Qi = []
nummer_of_threats = 10
losses = np.random.randint(1, 11, nummer_of_threats)
for i in range(nummer_of_threats):
    Qi.append(losses[i] / sum(losses))
    
print("Відносні величини збитків від реалізації загроз:")
np.round(Qi, 2)
```

```python
threat_df["Збитки від загрози"] = np.round(Qi, 2)

threat_df = threat_df.set_index("Назва загрози")
threat_df
```

**Для визначення загроз, які можуть викликати найбільші втрати, потрібно врахувати різні аспекти, такі як потенційні наслідки та ймовірність появи кожної загрози. Важливо також враховувати контекст і конкретні умови, в яких застосовується ця інформаційна система. Ось мої варіанти, які можуть вважатися топ-трьома загрозами з потенційними найбільшими втратами:**

► **Навмисне знищення інформації вірусами та ін.**:
   - Ця загроза може викликати серйозні втрати, оскільки віруси та інші шкідливі програми можуть нанести шкоду даним та системам, призводячи до їхнього втрати або пошкодження.
   - Імовірність появи цієї загрози може бути високою, оскільки шкідливі програми поширюються в мережі та можуть атакувати різні системи.

► **Відмова в обслуговуванні внаслідок дій зловмисника**:
   - Ця загроза може призвести до серйозних втрат часу та ресурсів, оскільки атаки на доступність можуть призвести до перерв у роботі системи або послуг.
   - Імовірність цієї загрози також може бути високою, оскільки атаки на доступність можуть бути відносно легкою і доступною для зловмисників.

► **Викрадення службової інформації з файлової системи зовнішніми зловмисниками**:
   - Ця загроза може призвести до великих втрат конфіденційної інформації, яка може бути використана для різних негативних цілей, включаючи шантаж або розкриття конфіденційної інформації.
   - Імовірність появи цієї загрози також може бути високою, якщо система не має належних захистів для захисту від несанкціонованого доступу.
