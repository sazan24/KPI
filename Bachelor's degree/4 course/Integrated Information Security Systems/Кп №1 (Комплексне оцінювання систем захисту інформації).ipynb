#### 1. Згенеруємо матрицю розмірністю **4х5х7** та заповнимо її по випадковому принципу значеннями q_ijk "експертних" оцінок:

```python
import numpy as np

# Розміри матриці
rows = 4
cols = 5
pages = 7

# Генеруємо матрицю A
matrix_A = np.random.uniform(0.2, 1, (pages, rows, cols)).round(2)

# Виводимо матрицю A
print(f"A) 0 <= q_ijk <= 1 \n{matrix_A}\n")

# Створюємо матрицю B, де значення 1, якщо відповідне значення в A > 0.5, і 0 в іншому випадку
matrix_B = (matrix_A > 0.5).astype(int)

# Виводимо матрицю B
print(f"B) q_ijk = {0, 1} \n{matrix_B}\n")

# Оператори умов для призначення значень відповідно до заданих діапазонів
conditions = [
    (matrix_A >= 0) & (matrix_A <= 0.3),
    (matrix_A > 0.3) & (matrix_A <= 0.5),
    (matrix_A > 0.5) & (matrix_A <= 0.7),
    (matrix_A > 0.7) & (matrix_A <= 0.9),
    (matrix_A > 0.9) & (matrix_A <= 1)
]

# Значення, що призначаються відповідно до умов
values = [1, 2, 3, 4, 5]

# Створюємо матрицю C, використовуючи функцію np.select
matrix_C = np.select(conditions, values)

# Виводимо матрицю C
print(f"C) q_ijk = {1, 2, 3, 4, 5} \n{matrix_C}")

# Збереження матриць у файлі для можливого їх перегляду та порівняння
matrices = [matrix_A, matrix_B, matrix_C]
filenames = ["matrix_A.txt", "matrix_B.txt", "matrix_C.txt"]
for matrix, filename in zip(matrices, filenames):
    with open(filename, 'w') as fp:
        print(matrix, file=fp)
```

#### 2. Обчислимо оцінку якості системи захисту інформації, яка описується відповідною матрицею:

##### 2.1. Для матриці C застосуємо звичайну формулу підрахунку середнього арифметичного ↓

```python
# Обчислюємо середнє арифметичне всіх значень матриці C
average_score = np.sum(matrix_C) / (rows * cols * pages)

print("Оцінка якості системи захисту інформації:", round(average_score, 2))
```

##### 2.2. Для матриці A застосуємо більш досконалу формулу, що враховуватиме вагу кожного елемента ↓

```python
# Задамо матрицю Q
Q = matrix_A

# Таблиця ваг
element_weights = [
    [0.2, 0.15, 0.15, 0.1, 0.05, 0.1, 0.25],
    [0.1, 0.2, 0.3, 0.4],
    [0.15, 0.25, 0.2, 0.1, 0.3]
]

# Розміри матриці Q
num_i, num_j, num_k = Q.shape

# Обчислюємо загальну оцінку
total_quality = 0.0
for k in range(num_k):
    for j in range(num_j):
        for i in range(num_i):
            total_quality += element_weights[2][k] * element_weights[1][j] * element_weights[0][i] * Q[i, j, k]

print("Загальна оцінка якості системи:", round(total_quality, 2))
```

#### 5. Обчислимо, які елементи матриці дають найгірший внесок у загальну якість КСЗІ?

```python
# Ініціалізуємо список для збереження найгірших значень внесків матриці
element_contributions = []

# Обчислюємо внесок кожного елементу
for k in range(num_k):
    for j in range(num_j):
        for i in range(num_i):
            element_contribution = element_weights[2][k] * element_weights[1][j] * element_weights[0][i] * Q[i, j, k]
            element_contributions.append((i, j, k, element_contribution))

# Сортуємо елементи за внеском у порядку зростання
sorted_element_contributions = sorted(element_contributions, key=lambda x: x[3])

# Обираємо п'ять елементів із найменшими внесками
top_5_contributions = sorted_element_contributions[:5]

print("Топ-5 елементів із найменшими внесками:")
for value in top_5_contributions:
    i, j, k, contribution = value
    print(f"• ({i}, {j}, {k}) має внесок {round(contribution, 5)}")
```
